# Fitting
## Intro
In this section, we will be fitting the data to several histograms to try to extract the hypothetical invariant mass of the Higgs boson.

```python
import ROOT
import numpy as np
import pandas as pd
import math
```

## Reading Data
This analysis is on the 2011 data.

```python
file_2011 = ROOT.TFile.Open("tree_TYPE_DATA_2011.root", "READ")
print(ROOT.TFile.GetListOfKeys(file_2011)[0])
```

#### TTree Object
Now that we have the name of the tree, which is `TYPE_DATA`, we can access the TTree object.

```python
t = file_2011.TYPE_DATA
```

#### Accessing TBranches
In order to perform analysis, we need the `invariant_mass` and `categ_Model10Categories` branches. We will select only these branches to increase read efficiency.

```python
t.SetBranchStatus("*", False)
t.SetBranchStatus("invariant_mass", True)
t.SetBranchStatus("categ_Model10Categories", True)
```

We will also load the category mapping file

```python
categ_mapping = pd.read_csv('categ_mapping.csv', delimiter=',')
print(categ_mapping)
```

## Visualization
We will first fit the background, then the residual, then combine the fits for a general fit.

```python
canv = ROOT.TCanvas("canv", "canv", 1400, 1600)

# Category Histograms
hist_arr = []
bg_hist_arr = []
for i in range(10):
    title = categ_mapping['categ_name'][i]
    bin_num = int(60/categ_mapping['bin_width_given'][i])
    print(title, bin_num)
    hist_arr.append(ROOT.TH1D(f"hist_{i}", r"{}".format(title), bin_num, 100, 160))
    hist_arr[i].GetXaxis().SetTitle(r"m_{\gamma\gamma} [GeV]")
    hist_arr[i].GetYaxis().SetTitle(f"Events / {categ_mapping['bin_width_given'][i]} GeV")
    hist_arr[i].Sumw2()
    hist_arr[i].SetStats(0)
    hist_arr[i].SetMarkerStyle(ROOT.kFullCircle)
    
    bg_hist_arr.append(ROOT.TH1D(f"bg_hist_{i}", r"{} (bg. fit)".format(title), bin_num, 100, 160))
    bg_hist_arr[i].GetYaxis().SetTitle(f"Events / {categ_mapping['bin_width_given'][i]} GeV")
    bg_hist_arr[i].Sumw2()
    bg_hist_arr[i].SetMarkerStyle(ROOT.kFullCircle)
    bg_hist_arr[i].SetStats(0)
    
total = ROOT.TH1D(f"total_hist_{i}", "Total", 30, 100, 160)
```

#### Filling Data
The next step is to fill the histograms with data read from the TBranches.

```python
l_lim = 120
r_lim = 132
for entry in range(0, t.GetEntries()):
    t.GetEntry(entry)
    inv_mass = t.invariant_mass
    total.Fill(inv_mass)
    categ = int(t.categ_Model10Categories)
    hist_arr[categ-1].Fill(inv_mass)
    if l_lim < inv_mass and inv_mass < r_lim:
        continue
    else:
        bg_hist_arr[categ-1].Fill(inv_mass)
```

#### Fitting Background
We chose the background function to be an exponential because it (experimentally) had the lowest chi2 value.

```python
class Background:
    """
    Wrapper to implement python function for ROOT TF1
    From ROOT documentation, 'expo' (builtin) is function for:
    exp([Constant] + [Slope]*x)
    which is what we have replicated
    PARAMETERS:
    par[0] = [Constant]
    par[1] = [Slope]
    """
    def __call__(self, arr, par):
        return math.exp(par[0] + par[1]*arr[0])
    
bg_func_arr = [ROOT.TF1(f"bg_func_{i}", Background(), 100, 160, 2) for i in range(len(bg_hist_arr))]
for func in bg_func_arr:
    func.SetParNames("Constant", "Slope")
    func.SetParameters(5,-1e-2)
```

```python
# Use fit function to fit each histogram
bg_fit_result = []
for i in range(len(bg_hist_arr)):
    bg_fit_result.append(bg_hist_arr[i].Fit(bg_func_arr[i], "SBM"))
    
for i in range(10):
    canv.Clear()
    canv.Draw()
    bg_hist_arr[i].Draw()
    canv.Print(f"bg_hist_folder//hist_{i}.png")
```

#### Fitting Signal
To fit the signal, we opted to use simply a Gaussian instead of the recommended CB and Gaussian linear combination because implementation of the CB and Gaussian linear combination was not functioning. We are choosing to float both the constant and the mean, as not all the datasets sseem to have a peak at the same location.

```python
class Signal:
    """
    Wrapper to implement python function for ROOT TF1.
    Function definition is a Gaussian.
    PARAMETERS:
    par[0]: gaus N
    par[5]: gaus m_h
    """
    
    def gaus(self, m_gg, N, m_h, sigma_g):
        k = (m_gg - m_h)/sigma_g
        return N * np.exp(-(k**2)/2)
    
    def __call__(self, arr, par):
        return self.gaus(arr[0], par[0],par[1],par[2])
    
sig_func_arr = [ROOT.TF1(f"sig_func_{i}", Signal(), 100, 160, 3) for i in range(len(hist_arr))]
```

Now we create the residual histograms. We do this by adding the signal, then subtracting the background function.

```python
resid_hist_arr = []

for i in range(10):
    title = categ_mapping['categ_name'][i]
    bin_num = int(60/categ_mapping['bin_width_given'][i])
    
    resid_hist_arr.append(ROOT.TH1D(f"resid_hist_{i}", r"{} (residual)".format(title), bin_num, 100, 160))
    resid_hist_arr[i].GetXaxis().SetTitle(r"m_{\gamma\gamma} [GeV]")
    resid_hist_arr[i].GetYaxis().SetTitle(f"Events / {categ_mapping['bin_width_given'][i]} GeV")
    resid_hist_arr[i].Sumw2()
    resid_hist_arr[i].SetStats(0)
    resid_hist_arr[i].SetMarkerStyle(ROOT.kFullCircle)
    resid_hist_arr[i].Add(hist_arr[i], 1)
    resid_hist_arr[i].Add(bg_func_arr[i], -1)
    
for i in range(10):
    canv.Clear()
    canv.Draw()
    resid_hist_arr[i].Draw()
    canv.Print(f"resid_hist_folder//hist_{i}.png")
```

```python
sig_fit_result = []
for i in range(len(resid_hist_arr)):
    sig_func_arr[i].FixParameter(2,categ_mapping["fwhm"][i]/2.355)
    sig_func_arr[i].SetParNames("Norm", "Mean", "Sigma")
    sig_func_arr[i].SetParLimits(0, 1e-1, 100)
    sig_func_arr[i].SetParLimits(1, l_lim, r_lim)
    sig_func_arr[i].SetParameter(0, 5)
    sig_func_arr[i].SetParameter(1, (r_lim+l_lim)/2)
    sig_fit_result.append(resid_hist_arr[i].Fit(sig_func_arr[i], "SBM"))
    
for i in range(10):
    canv.Clear()
    canv.Draw()
    resid_hist_arr[i].Draw()
    canv.Print(f"sig_hist_folder//hist_{i}.png")
```

```python
for i in range(len(hist_arr)):
    canv.Clear()
    pad1 = ROOT.TPad("pad1","pad1",0,0.3,1,1)
    pad1.SetBottomMargin(0)
    pad1.Draw()
    pad1.cd()
    bg_hist_arr[i].Draw()
    latex = ROOT.TLatex()
    latex.SetNDC()
    latex.SetTextSize(0.03)
    chi2 = bg_func_arr[i].GetChisquare()
    ndof = bg_func_arr[i].GetNDF()
    const = bg_func_arr[i].GetParameter(0)
    slope = bg_func_arr[i].GetParameter(1)
    latex.DrawText(0.5,0.85 , "[Constant] = %.1e"%(const))
    latex.DrawText(0.5,0.8 , "[Slope] = %.1e"%(slope))
    latex.DrawText(0.5,0.75,"chi2 / ndof = %.1e/%.1e = %.1e"%(chi2 , ndof , chi2 / ndof ))
    canv.cd()
    pad2 = ROOT.TPad("pad2","pad2" ,0,0,1,0.3)
    pad2.SetTopMargin(0)
    pad2.Draw()
    pad2.cd()
    resid_hist_arr[i].Draw()
    latex = ROOT.TLatex()
    latex.SetNDC()
    latex.SetTextSize(0.03)
    chi2 = sig_func_arr[i].GetChisquare()
    ndof = sig_func_arr[i].GetNDF()
    norm = sig_func_arr[i].GetParameter(0)
    mean = sig_func_arr[i].GetParameter(1)
    sigma = sig_func_arr[i].GetParameter(2)
    latex.DrawText(0.7,0.9 , "[Norm] = %.1e"%(norm))
    latex.DrawText(0.7,0.85 , "[Mean] = %.1e"%(mean))
    latex.DrawText(0.7,0.8 , "[Sigma] = %.1e"%(sigma))
    latex.DrawText(0.7,0.75,"chi2 / ndof = %.1e/%.1e = %.1e"%(chi2 , ndof , chi2 / ndof ))
    canv.Print(f"stack_hist_folder//hist_{i}.png")
```
